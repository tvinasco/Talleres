# 4. Solicite el radio (un numero) de un círculo e imprima como salida su diámetro, perímetro y su área
### Ejemplo
- ingreso: 3 
- imprime - Diametro: 6
- imprime - Perimetro: 18.84
- imprime - Area: 28.27

```python
from math import pi 

r = float(input("Escriba el radio: "))

diametro = 2*r

print ("El diametro del ciruclo es: {:.2f}".format(diametro))

perimetro = 2*pi*r
print ("El perimetro del ciruclo es: {:.2f}".format(perimetro))

area = pi*(r**2)

print ("El area del circulo es: {:.2f}".format(area))

```

# 9. Solicite la temperatura en grados Celsius e imprima en grados Kelvin y Fahrenheit 
### Ejemplo:
- ingreso: 38 (°C)
- imprime: 311.15 °K
- imprime: 100.4 °F

```python
# Fahrenheit = (9/5)*°C + 32 

C = float(input("Ingrese la temperatura en grados °C: "))

# Pasar de grados Celsius a Kelvin 
K = C + 273.15 
print(C, "°C equivale a", K, "°K")

# Pasar de grados Celsius a Fahrenheit
F = (9/5)*C + 32
print(C, "°C equivale a", F, "°F")
```

# 12. Solicite una distancia (metros) y la velocidad (metros/segundos) e imprima el tiempo que se demora el recorrido
### Ejemplo:
- ingreso: 150 (metros)
- ingreso: 15 (metros/segundo)
- Imprime: 10 segundos

```python
# Velocidad = Distancia * Tiempo
# Tiempo = Distancia / Velocidad


d = float(input("Ingrese la distancia en metros a recorrer?: "))
v = float(input("Ingrese la Velocidad que alcanzo metros/segundos: "))

t = d/v

print("El recorrido tardara un tiempo de {:.0f}".format(t), "segundos")
```
